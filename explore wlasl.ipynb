```python
import json
import itertools
file_path = '../WLASL-master/start_kit/WLASL_v0.3.json'

num_class = 100 # to use
with open(file_path) as ipf:
    content = json.load(ipf)

cnt_train = 0
cnt_val = 0
cnt_test = 0

for ent in content:
    gloss = ent['gloss']

    for inst in ent['instances']:
        split = inst['split']

        if split == 'train':
            cnt_train += 1
        elif split == 'val':
            cnt_val += 1
        elif split == 'test':
            cnt_test += 1
        else:
            raise ValueError("Invalid split.")

print('total glosses: {}'.format(len(content)))
print('total samples: {}'.format(cnt_train + cnt_val + cnt_test))
```

```python
class_dict = dict()
for c in content:
    class_dict[c['gloss']] = len(c['instances'])
class_dict
```

```python
sorted_clas_dict = {k: v for k, v in sorted(class_dict.items(), key=lambda item: item[1])}
reverse =dict(reversed(list(sorted_clas_dict.items())))
# select the ones with most samples
sorted_clas_dict = dict(itertools.islice(reverse.items(), num_class))
print(sorted_clas_dict)
select_classes = sorted_clas_dict.keys()
print('\n\nselect_classes:', select_classes)
```

```python
files_to_process = []

for c in content:
    if c['gloss'] not in select_classes:
        continue
    for i in c['instances']:
        files_to_process.append(i['video_id'])
print('Total num. of files to process:', len(files_to_process))
```

```python
savepath = '/mnt/HDD04/WASL Dataset/Scripts/'
with open(savepath + 'files' + str(num_class) + '.txt', 'w') as f:
    for item in files_to_process:
        f.write(item+'\n')
```
